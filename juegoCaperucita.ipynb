```python
import time
from IPython.display import clear_output
import ipywidgets as widgets
from IPython.display import display

VIKINGO, LOBO, CAPERUCITA, UVAS, BARCO = 0, 1, 2, 3, 4
estado_inicial = ["L", "L", "L", "L", "L"]
personajes = ['ü™ì', 'üê∫', 'üëß', 'üçá']

def mostrar_estado(estado):
    izquierda = [personajes[i] for i in range(4) if estado[i] == "L"]
    derecha = [personajes[i] for i in range(4) if estado[i] == "R"]
    barco = "Izquierda" if estado[BARCO] == "L" else "Derecha"
    display(widgets.HTML(f"<b>[{', '.join(izquierda)}] üåäüåäüåäüåäüåä [{', '.join(derecha)}]</b>"))
    display(widgets.HTML(f"<b>El barco est√° en la orilla: {barco}</b>"))

def verificar_reglas(estado):
    if estado[LOBO] == estado[CAPERUCITA] and estado[VIKINGO] != estado[LOBO]:
        return False, "El lobo se comi√≥ a Caperucita."
    if estado[CAPERUCITA] == estado[UVAS] and estado[VIKINGO] != estado[CAPERUCITA]:
        return False, "Caperucita se comi√≥ las uvas."
    return True, ""

def mover(estado, eleccion):
    nuevo_estado = estado[:]
    nuevo_estado[VIKINGO] = nuevo_estado[BARCO] = "R" if estado[BARCO] == "L" else "L"
    if eleccion != "solo":
        nuevo_estado[eleccion] = nuevo_estado[BARCO]
    return nuevo_estado

def obtener_opciones(estado):
    opciones = ["solo"]
    for i in [LOBO, CAPERUCITA, UVAS]:
        if estado[i] == estado[BARCO]:
            opciones.append(i)
    return opciones

def actualizar_opciones(estado):
    opciones = obtener_opciones(estado)
    botones = []
    for opcion in opciones:
        nombre = ["solo", "al lobo", "a Caperucita", "a las uvas"][opcion] if isinstance(opcion, int) else opcion
        boton = widgets.Button(description=f"Cruzar {nombre}")
        boton.on_click(lambda b, opcion=opcion: procesar_opcion(estado, opcion))
        botones.append(boton)
    return widgets.VBox(botones)

def animar_cruce(estado, eleccion):
    direccion = "derecha" if estado[BARCO] == "L" else "izquierda"
    for i in range(6, 0, -1) if direccion == "derecha" else range(1, 7):
        clear_output(wait=True)
        izquierda = [personajes[i] for i in range(4) if estado[i] == "L"]
        derecha = [personajes[i] for i in range(4) if estado[i] == "R"]
        barco = "Izquierda" if estado[BARCO] == "L" else "Derecha"
        print(f"[{', '.join(izquierda)}] {'üåä' * (6-i if direccion == 'derecha' else 6-i)}üõ∂{'üåä' * (i if direccion == 'derecha' else i)} [{', '.join(derecha)}]")
        print(f"El barco est√° en la orilla: {barco}")
        time.sleep(0.5)

def procesar_opcion(estado, eleccion):
    animar_cruce(estado, eleccion)
    nuevo_estado = mover(estado, eleccion)
    jugar(nuevo_estado)

def jugar(estado=None):
    if estado is None:
        estado = estado_inicial[:]
    clear_output(wait=True)
    mostrar_estado(estado)
    valido, mensaje = verificar_reglas(estado)
    if not valido:
        display(widgets.HTML(f"<b>{mensaje}</b>"))
    elif estado[1:] == ["R", "R", "R", "R"]:
        display(widgets.HTML("<b>GANASTE! Todos lograron cruzar el r√≠o con √©xito!.</b>"))
    else:
        display(widgets.HTML("<b>¬øQu√© quieres hacer?</b>"))
        display(actualizar_opciones(estado))

jugar()
```
